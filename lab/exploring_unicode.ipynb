```python
from sys import maxunicode
maxunicode
```

```python
import unicodedata

for code in range(32, 10000):
    char = chr(code)
    name = unicodedata.name(char, '')
    if name:
        print(f'U+{code:04X}', char, name, sep='\t')
```

```python
for code in range(32, maxunicode):
    char = chr(code)
    name = ' ' + unicodedata.name(char, '') + ' '
    if ' CAT ' in name:
        print(f'U+{code:04X}', char, name, sep='\t')
```

```python
for code in range(32, maxunicode):
    char = chr(code)
    name = unicodedata.name(char, '')
    if name:
        words = name.split()
        if 'CAT' in words:
            print(f'U+{code:04X}', char, name, sep='\t')
```
